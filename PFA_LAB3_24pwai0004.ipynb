Variable Length Arguements(*args)

```python
def total_numbers(*numbers):
    return sum(numbers)
print("Total is this: ",total_numbers(1,2,3))
print("Total is this: ",total_numbers(2,3))
```

**Keyword Variable Arguements(kwargs)**

```python
def student_info(**data):
    for k,v in data.items():
        print(f"{k}:{v}")
student_info(name="Warisha", age=20, major='Artficial Intelligence')
```

**Lambda Function**

```python
sq=lambda x:x*x
print("Squared:",sq(5))
```

**Lambda with List**

```python
square=lambda x:x*x
numbers=[1,2,3,4,5]
squared_numbers=list(map(square,numbers))
print(squared_numbers)
```

**Lambda with filter**

```python
marks=[45,67,82,30,90,55]
greater_than_50=list(filter(lambda x:x>50,marks))
print(greater_than_50)
```

Students will build a small program that:
 stores student marks
 calculates total and average
 finds highest marks
 uses functions and lambda
Task Description
Students will create a program that:
1. Takes student marks as input
2. Uses functions to process data
3. Calculates total marks
4. Calculates average marks
5. Finds highest mark using lambda

```python
def calculate_total(marks_list):
    return sum(marks_list)
def calculate_average(total, count):
    return total/count if count > 0 else 0

print("Enter student marks separated by spaces: ")
marks=list(map(float, input(">").split()))
```
