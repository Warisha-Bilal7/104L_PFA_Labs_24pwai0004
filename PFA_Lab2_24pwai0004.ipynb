**Mini Calculator**

```python
#Ask user for two numbers (use float() so decimals work).
num1=float(input("Please enter the first number: "))
num2=float(input("Please enter the second number: "))
#Perform calculations and handle results
result=None
valid=True
#Ask for operation: + - * / ** % (as string)
operation=input("Please enter the operation(+,-.*,/,**,%): ")
#Calculate and print result using f-string, formatted nicely upto 2 decimal places
if operation=="+":
  result=num1 + num2
elif operation=="-":
  result=num1 - num2
elif operation=="*":
  result=num1 * num2
elif operation == "/":
  if num2 !=0:
    result= num1/num2
  else:
    print("Error: Division by zero is not allowed.")
    valid=False
elif operation=="**":
  result=num1 ** num2
elif operation=="%":
  result=num1 % num2
else:
  print("Error: Invalid operation.")
  valid=False
#Print results formatted to 2 decimal places
if valid:
  print(f"The result is: {result:.2f}")
```

**Challenge A – Even/Odd &amp; Multiple Checker (while loop)**

```python
#Ask user for starting and ending point
start=int(input("Please enter the starting point: "))
end=int(input("Please enter the ending point: "))
num=start
while num<=end:

  #Skip printing the number itself if its divisible by 7 (use continue)
  if num %7==0:
    num += 1
    continue
  #If number is even and divisible by 5, print ("Even and a Multiple of 5")
  if num %2==0 and num %5==0:
    print(f"{num} is even and a mutilple of 5")
  #If number is divisible by 2, it is even
  elif num % 2 ==0:
    print(f"{num} is even")
  #if number is divisible by 5, it is odd
  elif num % 5 ==0:
    print(f"{num} is a multiple of 5")
  else:
    print(num)
#increment number until ending point is reached
num +=1
```

**Password Strength Checker (if-elif nested)**

```python
password=input("Enter a password: ")
if len(password) < 6:
  print("Too short!")
else:
  has_digit=any(char.isdigit() for char in password)
  has_uppercase=any(char.isupper() for char in password)
  if not has_digit:
    print("Add at least one number.")
  elif not has_uppercase:
    print("Add at least one capital letter.")
  else:
    print("Strong password-good choice!")
```
