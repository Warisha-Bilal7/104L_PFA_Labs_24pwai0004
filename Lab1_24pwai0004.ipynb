Grade System

```python
marks = int(input("Enter your marks: ")) #user iNput

if marks >= 80:
    print("Grade: A")
elif marks >= 60:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Result: Fail")
```

Function

```python
# Function to calculate average
def average(a, b, c):
    avg = (a + b + c) / 3
    return avg

result = average(80, 75, 90)
print("Average is:", result)
```
